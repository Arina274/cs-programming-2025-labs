```python
def f(v, fr, to):
    d = {'s': 1, 'm': 60, 'h': 3600}
    return v * d[fr] / d[to]

print(f(4, 'h', 'm'))
print(f(30, 'm', 'h'))
print(f(12, 's', 'h'))
```

Задание 2

```python
def f(x: int, y: int):
    if x < 30000:
        return 0
    base = min(0.3 * (x // 10000), 5)
    total = 0  
    for i in range(y):
        if i <= 2:
            rate = 3
        elif i < 6:
            rate = 5
        else:
            rate = 2
        g = x * ((base + rate) * 0.01)
        total += g
        x+=g
    return total

print(f(30000, 3))
print(f(100000, 5))
print(f(200000, 8))
```

Задание 3

```python

def f(n):
    if n<2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

def t(start, end):
    if start > end:
        return "Error! Начало диапазона больше конца"
    if end < 2:
        return "Error! В диапазоне нет простых чисел"
    
    primes = []
    for i in range(start, end + 1):  
        if f(i):
            primes.append(i)
    return " ".join(map(str, primes))
print(t(1, 10))     
print(t(15, 120))   
print(t(0, 1))     
print(t(10, 1))    
print(t(-5, 1))  
```

Задание 4

```python
def matrix(n):
    return [list(map(int, input().split())) for _ in range(n)]
def summa(x, y,n):
    for i in range(n):
        for j in range(n):
            print(x[i][j] + y[i][j], end=" ")
        print()
n = int(input())
first = matrix(n)
second = matrix(n)
if any(len(row) != n for row in first + second):
    print("Error!")
else:
    summa(first, second, n)
```

```python
def f(text):
    s = ''.join(c for c in text.lower() if c.isalnum())
    return "Да" if s == s[::-1] else "Нет"

text = input()
print(f(text))
```
