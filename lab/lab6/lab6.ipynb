Задание 8

```python
import random
v=['камень','ножницы','бумага','ящерица','спок']
x=input("Выбери:камень,ножницы,бумага,ящерица,спок")
c= random.choice(v)
print(c,'- рандомный выбор')
if x==c:
    print('ничья')
elif (x=='камень' and(c=='ножницы' or c=='ящерица')) or \
    (x=='ножницы' and(c=='бумага' or c=='ящерица')) or \
    (x=='бумага' and(c=='камень' or c=='спок')) or \
    (x=='ящерица' and(c=='спок' or c=='бумага')) or \
    (x=='спок' and(c=='камень' or c=='ножницы')):
    print('Ты выиграл')
else:
    print('Компьютер выиграл')
```

Задание 9

```python
words=["яблоко", "груша", "банан", "киви", "апельсин", "ананас"]
d={}
for w in words:
    one=w[0]
    if one not in d:
        d[one]=[]
    d[one].append(w)
print(d)
```
